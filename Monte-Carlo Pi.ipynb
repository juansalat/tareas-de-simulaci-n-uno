```python
#Método de Monte-Carlo para aproximar el valor de Pi.
"""
Importamos el módulo random para generar números aleatorios. Definimos la variable
para el número de números aleatorios que vamos a generar. Contamos los que ocurren
dentro del círculo unitario. Utilizamos un for para generarlos, y verificamos con
un if, si es que están dentro del círculo, si ocurre que sí, entonces incrementamos
el valor de c en uno. Y al final vamos a imprimir el valor aproximado de Pi. Cuatro
veces la proporción de puntods dentro del círculo respecto a n. 
"""
import random
n=10**6
c=0
for i in range(n):
    if random.random()**2+random.random()**2<1:
        c=c+1
print("El valor aproximado de Pi=",4*c/n)
hola="Hola"
def plottingpi():
    import matplotlib.pyplot as plt
    c=0
    x_in=[]
    y_in=[]
    x_out=[]
    y_out=[]
    n=1000
    for i in range(n):
        x=random.random()
        y=random.random()
        if x**2+y**2<1:
            c+=1
            x_in.append(x)
            y_in.append(y)
        else:
            x_out.append(x)
            y_out.append(y)
    pi=4*c/n
    plt.scatter(x_in,y_in,color='blue',s=1,label='Inside')
    plt.scatter(x_out,y_out,color='red',s=1,label='Outside')
    plt.gca().add_artist(plt.Circle((0,0),1,color='black',fill=False))
    plt.gca().set_aspect('equal','box')
    plt.title('Monte Carlo Estimation of Pi')
    plt.xlabel('X Axis')
    plt.ylabel('Y Axis')
    plt.legend()
    plt.show()
    return hola
plottingpi()
```
