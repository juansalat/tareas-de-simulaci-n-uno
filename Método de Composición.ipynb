```python
"""
Método de Composición.

Definimos las funciones de densidad de probabilidad de las dos distribuciones normales.
Y tenemos que asignar probabilidades a las distribuciones. Generamos muestras, según la
probabilidad de cada una de las distribuciones será la muestra que se vaya a generar. 
Generaremos un número específico de muestras. Y graficamos. 
"""
import numpy as np
import matplotlib.pyplot as plt
def normal_distribution(x, mu, sigma):
    return (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
mu1, sigma1 = -2, 0.5
mu2, sigma2 = 2, 1
p1, p2 = 0.3, 0.7
def composition_sampling(mu1, sigma1, p1, mu2, sigma2, p2, sample_size):
    samples = []
    for i in range(sample_size):
        if np.random.rand() < p1:
            samples.append(np.random.normal(mu1, sigma1))
        else:
            samples.append(np.random.normal(mu2, sigma2))
    return np.array(samples)
sample_size = 10000
samples = composition_sampling(mu1, sigma1, p1, mu2, sigma2, p2, sample_size)
x = np.linspace(-6, 6, 1000)
plt.hist(samples, bins=50, density=True, alpha=0.5, label='Muestras')
plt.plot(x, p1 * normal_distribution(x, mu1, sigma1) + p2 * normal_distribution(x, mu2, sigma2), label='Distribución objetivo')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.legend()
plt.title('Muestras generadas por el método de composición')
plt.show()
print("Función de Distribución Acumulativa Empírica")
def distribucion_acumulativa_empirica(muestras, x):
    return np.array([np.sum(muestras <= xi) / len(muestras) for xi in x])
print("Gráfica de la Función de Distribución Acumulativa") 
fda_empirica = distribucion_acumulativa_empirica(samples, x)
plt.plot(x, fda_empirica, label='FDA empírica')
plt.xlabel('x')
plt.ylabel('Probabilidad acumulada')
plt.legend()
plt.title('Función de Distribución Acumulativa (FDA)')
plt.show()
print("Histograma de las muestras")
plt.hist(samples, bins=50, color='blue', edgecolor='black')
plt.xlabel('x')
plt.ylabel('Frecuencia')
plt.title('Histograma de las muestras generadas')
plt.show()
```
