```python
#Método de la Transformada Inversa. 
"""
Éste programa genera muestras aleatorias de una distribución exponencial
utilizando el método de la transformada inversa. Tomamos u de una distribución
uniforme en el intervalo [0,1] y aplicamos la transformada inversa para obtener
una muestra de una distribución exponencial. log(1/(1-u)) es la inversa de la
función de distribución acumulativa de una distribución exponencial con parámetro
lambda igual a uno. Generamos diez mil muestras de una distribución uniforme
en el intervalo [0,1]. Transformamos las muestras uniformes a exponenciales.
Y graficamos. 
"""
import numpy as np
import matplotlib.pyplot as plt
def distribucion_inversa(u):
    return np.log(1/(1-u))
tamano_muestra=10000
uniformes=np.random.uniform(0,1,tamano_muestra)
muestras=distribucion_inversa(uniformes)
x=np.linspace(0,10,1000)
plt.hist(muestras,bins=50,density=True,alpha=0.5,label='Muestras')
plt.plot(x,np.exp(-x),label='Distribución objetivo')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.legend()
plt.title('Muestras generadas por el método de la transformada inversa')
plt.show()
print("Función de Distribución Acumulativa Empírica.")
def distribucion_acumulativa_empirica(muestras, x):
    return np.array([np.sum(muestras <= xi) / len(muestras) for xi in x])
print("Gráfica de la Función de Distribución Acumulativa")
fda_empirica = distribucion_acumulativa_empirica(muestras, x)
plt.plot(x, fda_empirica, label='FDA empírica')
plt.xlabel('x')
plt.ylabel('Probabilidad acumulada')
plt.legend()
plt.title('Función de Distribución Acumulativa (FDA)')
plt.show()
print("Histograma de las muestras")
plt.hist(muestras, bins=50, color='blue', edgecolor='black')
plt.xlabel('x')
plt.ylabel('Frecuencia')
plt.title('Histograma de las muestras generadas')
plt.show()
```
